**Задача 1.** Когда используется критерий Стьюдента, а когда Z –критерий?


Критерий Стьюдента используется когда неизвестна дисперсия генеральной совокупности и размер выборки небольшой (n < 30). 

Z - критетрий используется когда размер выборки большой (n  > 30) и дисперсия генеральной совокупности известна.

**Задача 2.**

Проведите тест гипотезы. Утверждается, что шарики для подшипников, изготовленные
автоматическим станком, имеют средний диаметр 17 мм.
Используя односторонний критерий с α=0,05, проверить эту гипотезу, если в выборке из
n=100 шариков средний диаметр
оказался равным 17.5 мм, а дисперсия известна и равна 4 кв. мм.

Принимаем гипотезу Н0 - диамет шариков = 17 мм

Н1 - диамет шариков > 17 мм


```python
x_average, n, mu, disp = 17, 100, 17.5, 4 

z = (mu - x_average) / (disp**(1/2) / (n**(1/2)))
z
```

По таблице для α = 0.05: z_крит = 1.645

Поскольку z > z_крит, отвергаем нулевую гипотезу H0 и принимаем альтернативную гипотезу H1. 

Т.е. шарики для подшипников, изготовленные автоматическим станком, ***имеют средний диаметр больше 17 мм.***

**Задача 3.**

Проведите тест гипотезы. Продавец утверждает, что средний вес пачки печенья
составляет 200 г.
Из партии извлечена выборка из 10 пачек. Вес каждой пачки составляет:
202, 203, 199, 197, 195, 201, 200, 204, 194, 190.
Известно, что их веса распределены нормально.
Верно ли утверждение продавца, если учитывать, что доверительная вероятность равна
99%? (Провести двусторонний тест.)

Принимаем гипотезу Н0 - средний вес пачки печенья = 200 г

Н1 - средний вес пачки печенья =! 200 г

Будем использовать критерий Стьюдента с учетом уровня значимости α/2 и числа степеней свободы k = n-1:


```python
mu, alpha, n = 200, 0.01/2, 10
k = n - 1
weight = [202, 203, 199, 197, 195, 201, 200, 204, 194, 190]
x_average = sum(weight) / len(weight)

temp = 0
for i in range(len(weight)):
    temp += (weight[i] - x_average)**2
sko = (temp / k)**(1/2)
t = round((x_average - mu) / (sko/ n**(1/2)),3)

print(f't-статистика = {t}')
print(f'Число степеней свободы "k" = {k}')
```

Критическое значение при α/2 = 0.005 и числа степеней свободы k = 9: t_крит = 3.3250

Поскольку |t| < t_крит, принимается гипотеза Н0, т.е. ***продавец прав, если утверждает, что средний вес пачки печенья составляет 200 г.*** 

**Задача 4.**

Есть ли статистически значимые различия в росте
дочерей?

Рост матерей 172, 177, 158, 170, 178,175, 164, 160, 169, 165
Рост взрослых дочерей: 173, 175, 162, 174, 175, 168, 155, 170, 160, 163

Принимает уровень значимости α = 0.05. 

Нулевая гипотеза H0: средний рост матерей и дочерей одинаковый, тогда альтернативная гипотеза H1: средний рост матерей и дочерей различается.

Используем ***двухвыборочный тест с независимыми выборками.***

```python
import scipy.stats as stats

mothers_height = [172, 177, 158, 170, 178,175, 164, 160, 169, 165]
daughters_height = [173, 175, 162, 174, 175, 168, 155, 170, 160, 163]

rez = stats.ttest_ind(mothers_height, daughters_height)
rez
```

Поскольку pvalue > α, принимаем гипотеззу Н0, т.е. нет статистически значимых различий в росте.
